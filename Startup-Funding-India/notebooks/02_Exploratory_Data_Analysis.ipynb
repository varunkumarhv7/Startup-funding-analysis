```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv("../data/cleaned/cleaned_startup_funding.csv")
```

```python
df.head()
```

```python
df["Industry"].value_counts().head(10)
```

```python
top_industries = df["Industry"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_industries.plot(kind="bar")
plt.title("Top 10 industries")
plt.xlabel("Industry")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_cities = df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_cities.plot(kind="bar")
plt.title("Top 10 Startup Cities")
plt.xlabel("City")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_investors = df["Investors"].value_counts().head(10)

plt.figure(figsize=(12,5))
top_investors.plot(kind="bar")
plt.title("Top 10 Investors")
plt.xlabel("Investor")
plt.ylabel("Number of Investments")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
industry_funding = (
    df.groupby("Industry")["InvestmentAmount_USD"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))
industry_funding.plot(kind="bar")
plt.title("Top 10 Industries by Total Funding")
plt.xlabel("Industry")
plt.ylabel("Funding (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
city_funding = (
    df.groupby("City")["InvestmentAmount_USD"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))
city_funding.plot(kind="bar")
plt.title("Top 10 Cities by Funding")
plt.xlabel("City")
plt.ylabel("Funding (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df["Year"] = pd.to_datetime(df["Date"]).dt.year
```

```python
yearly_funding = df.groupby("Year")["InvestmentAmount_USD"].sum()

plt.figure(figsize=(10,5))
yearly_funding.plot(marker="o")
plt.title("Funding Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Funding (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
